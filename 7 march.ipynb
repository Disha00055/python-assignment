Q1. What are the three measures of central tendency?
answer: The three measures of central tendency are:

Mean: The mean is the most commonly used measure of central tendency. It is calculated by summing up all the values in a dataset and dividing by the total number of values. The mean represents the average or the arithmetic center of the data.

Median: The median is the middle value in a sorted dataset. To find the median, the data is arranged in ascending or descending order, and the value that separates the higher and lower halves of the dataset is identified. The median is particularly useful when dealing with skewed distributions or datasets with outliers.

Mode: The mode is the value or values that appear most frequently in a dataset. Unlike the mean and median, which are numerical calculations, the mode is based on the frequency of values. A dataset can have no mode (if no value is repeated), one mode (unimodal), two modes (bimodal), or more than two modes (multimodal). The mode is useful for identifying the most common or typical values in a dataset.

These three measures of central tendency provide different perspectives on the center or typical value of a dataset and are appropriate in different situations. The mean is sensitive to extreme values, the median is resistant to outliers, and the mode emphasizes the most frequently occurring values.



Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?
answer:The mean, median, and mode are three measures of central tendency used to describe the center or typical value of a dataset. Here are the differences between these measures and how they are used:

Mean:

Definition: The mean, also known as the average, is calculated by summing up all the values in a dataset and dividing by the total number of values.
Calculation: Mean = (Sum of all values) / (Total number of values).
Sensitivity to Outliers: The mean is sensitive to extreme values or outliers because it takes into account the magnitude of all values.
Usage: The mean is widely used when the dataset has a symmetric distribution or when the goal is to summarize the overall average of the data.
Median:

Definition: The median is the middle value in a sorted dataset, separating the higher and lower halves of the data. If the dataset has an even number of values, the median is the average of the two middle values.
Calculation: Sort the dataset and find the middle value(s).
Resistance to Outliers: The median is resistant to outliers because it focuses on the position of values rather than their magnitude.
Usage: The median is commonly used when dealing with skewed distributions or datasets that contain outliers. It provides a robust measure of central tendency.
Mode:

Definition: The mode represents the value(s) that appear most frequently in a dataset.
Calculation: Identify the value(s) with the highest frequency.
Handling Non-Numeric Data: The mode can be used with both numerical and categorical data.
Usage: The mode is useful for identifying the most common or typical values in a dataset. It is often used when working with categorical or discrete data, such as survey responses or categorical variables.
These measures of central tendency serve different purposes and are chosen based on the characteristics of the dataset and the goals of the analysis. The mean provides the average value, the median represents the middle value, and the mode highlights the most frequently occurring value(s). By using these measures appropriately, you can gain insights into the center or typical value of the data.
