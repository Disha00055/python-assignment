**Exception in Python:**
An Exception in Python is an event that occurs during the execution of a program, disrupting the normal flow of the program's instructions. Exceptions are raised when an error or unexpected condition occurs that needs to be handled. They allow for graceful error handling and recovery in a program.

**Difference between Exceptions and Syntax Errors:**
1. **Exceptions:**
   - Exceptions occur during the runtime execution of a program.
   - They can be caused by factors such as invalid input, file I/O errors, division by zero, and more.
   - Exceptions are instances of classes that inherit from the `BaseException` class.
   - They can be caught and handled using `try` and `except` blocks.
   - Examples of common exceptions include `ValueError`, `TypeError`, and `FileNotFoundError`.

2. **Syntax Errors:**
   - Syntax errors are detected by the Python interpreter during the parsing phase before a program is executed.
   - They are caused by violations of the Python language syntax rules.
   - Syntax errors prevent the program from running and need to be fixed before execution.
   - Examples of syntax errors include missing colons, unmatched parentheses, and incorrect indentation.

In summary, exceptions are runtime errors that can be caught and handled, while syntax errors are detected before program execution and require fixing before the program can run.

Q2. What happens when an exception is not handled? Explain with an example
answer:When an exception is not handled in a program, it leads to what is known as an "unhandled exception." An unhandled exception occurs when an error occurs during program execution, but there is no code in place to catch and handle that specific type of exception. As a result, the program's normal execution is abruptly terminated, and an error message is displayed, providing information about the unhandled exception.

Here's an example to illustrate this:

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num  # This can potentially raise a ZeroDivisionError
except ValueError:
    print("Invalid input! Please enter a valid number.")
```

In this example, the program tries to get a number from the user and then calculates the result of dividing 10 by the entered number. If the user enters a valid number, everything works fine. However, if the user enters `0`, a `ZeroDivisionError` will occur because you cannot divide by zero.

If the program runs into a `ZeroDivisionError` and there's no `except` block to handle it, the exception becomes unhandled. Here's what happens:

1. The program encounters the `ZeroDivisionError` during the division operation.
2. Since there's no `except ZeroDivisionError:` block to catch and handle this exception, the program's normal execution is abruptly halted.
3. An error message is displayed, indicating the type of exception (`ZeroDivisionError`) that occurred and a traceback of where the error occurred in the code.

The error message might look something like this:
```
ZeroDivisionError: division by zero
```

In this scenario, the program terminates unexpectedly due to an unhandled exception. To prevent this, it's important to include appropriate `try` and `except` blocks to handle exceptions that could potentially occur during program execution. This allows the program to gracefully handle errors and continue executing, even if exceptions occur.


Q3. Which Python statements are used to ,catch and handle exceptions? Explain with
an example
answer:
In Python, the `try`, `except`, `else`, and `finally` statements are used to handle exceptions. These statements provide a structured way to manage errors and control the flow of a program when exceptions occur.

Here's a breakdown of how these statements work along with an example:

- **try:** The `try` block contains the code that might raise an exception. It's the part of the code where you anticipate that an exception might occur.

- **except:** The `except` block is used to catch and handle exceptions that occur within the associated `try` block. You can specify the type of exception you want to catch, or you can catch a generic `Exception` to handle all types of exceptions.

- **else:** The `else` block is optional and is executed if no exceptions are raised in the `try` block. It's often used to define code that should run when no exceptions occur.

- **finally:** The `finally` block is also optional and is executed regardless of whether an exception occurred or not. It's used to define code that must be executed, such as cleanup operations, regardless of the exception handling.

Here's an example illustrating how these statements are used:

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num  # This can potentially raise a ZeroDivisionError
except ValueError:
    print("Invalid input! Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print("Division result:", result)
finally:
    print("Execution completed.")

print("Program continues...")
```

In this example:
- The `try` block attempts to get a number from the user and perform a division.
- If the user enters an invalid value, a `ValueError` is caught and handled.
- If the user enters `0`, a `ZeroDivisionError` is caught and handled.
- If no exceptions occur, the `else` block prints the division result.
- The `finally` block always prints "Execution completed," indicating the end of the try-except block.
- The program continues executing regardless of whether an exception occurred or not.

This structured approach to handling exceptions allows you to provide meaningful error messages, manage program flow, and perform necessary cleanup actions, enhancing the robustness and reliability of your code.

Q5. What are Custom Exceptions in python? What do we need Custom Exceptions? Explain with an example

Custom exceptions in Python are user-defined exceptions that allow you to create your own types of exceptions to handle specific situations in your code. While Python provides a variety of built-in exceptions (like `ValueError`, `TypeError`, etc.), there are cases where you might want to create your own exceptions to make your code more understandable and maintainable. Custom exceptions help in raising and handling errors that are specific to your application's logic.

Here's why you might need custom exceptions:

1. **Clarity and Readability**: By defining custom exceptions, you can give meaningful names to errors that are specific to your domain or application. This improves the readability of your code and helps other developers understand the nature of the error.

2. **Modularity**: Custom exceptions allow you to encapsulate error handling logic in a separate part of your code. This enhances modularity and separation of concerns.

3. **Debugging**: When an error occurs, a well-named custom exception provides more context about what went wrong, making debugging easier.

Let's illustrate this with an example:

```python
class BankAccountError(Exception):
    """Base class for bank account related exceptions."""
    pass

class InsufficientFundsError(BankAccountError):
    """Raised when a withdrawal is attempted with insufficient funds."""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds. Balance: {balance}, Withdrawal amount: {amount}")

class NegativeAmountError(BankAccountError):
    """Raised when a negative amount is used for a transaction."""
    def __init__(self, amount):
        self.amount = amount
        super().__init__(f"Negative amount not allowed: {amount}")

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount < 0:
            raise NegativeAmountError(amount)
        
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        
        self.balance -= amount
        return self.balance

try:
    account = BankAccount(1000)
    withdrawal_amount = -200  # Negative amount
    new_balance = account.withdraw(withdrawal_amount)
except BankAccountError as e:
    print("Error:", e)
else:
    print("Withdrawal successful. New balance:", new_balance)
```

In this example, we define three custom exceptions related to bank account operations: `InsufficientFundsError`, `NegativeAmountError`, and `BankAccountError`. Each custom exception inherits from the base class `BankAccountError`.

The `BankAccount` class uses these custom exceptions to handle various scenarios such as insufficient funds and negative withdrawal amounts. If a negative withdrawal amount is attempted, a `NegativeAmountError` is raised. If a withdrawal is attempted with insufficient funds, an `InsufficientFundsError` is raised.

By using custom exceptions, the code becomes more organized, and it's easier to understand the specific errors that might occur during bank account operations.

Certainly! Here's an example of creating a custom exception class and using it to handle an exception:

```python
class CustomError(Exception):
    """Custom exception class."""
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def example_function(value):
    if value < 0:
        raise CustomError("Value cannot be negative.")

try:
    user_input = int(input("Enter a positive number: "))
    example_function(user_input)
except CustomError as ce:
    print("Custom Error:", ce)
else:
    print("No custom error occurred.")
finally:
    print("Exception handling complete.")
```

In this example:

1. We define a custom exception class `CustomError` that inherits from the built-in `Exception` class. The `__init__` method is used to initialize the exception instance with a custom error message.

2. The `example_function` takes a value as an argument and raises a `CustomError` if the value is negative.

3. In the `try` block, the user is prompted to input a number. If the number is negative and the `example_function` is called, a `CustomError` is raised.

4. The `except` block catches the `CustomError` and prints the custom error message.

5. The `else` block is executed if no custom error occurs.

6. The `finally` block is executed regardless of whether an exception occurred or not.

When you run this code and input a negative number, the custom exception will be raised and caught, displaying the custom error message. If you input a positive number, the program will execute the `else` block. Regardless of the input, the `finally` block will be executed.
