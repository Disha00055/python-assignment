Q14.how is memory managed in python.
ans:Memory management in Python is handled by the Python interpreter itself, using a combination of techniques, including:

Reference counting: Python keeps track of how many references each object has and frees the memory associated with an object when it has no more references. This is the primary memory management technique used in Python.

Garbage collection: Python also has a garbage collector that periodically runs to find and free objects that are no longer referenced. The garbage collector is used to handle cyclic references, where objects reference each other in a loop and no external references exist.

Memory pooling: Python uses memory pooling to reduce the overhead of memory allocation and deallocation. Memory pooling involves reusing memory blocks instead of freeing and reallocating them.

Overall, Python's memory management system is designed to be efficient and automatic, so that developers don't need to worry about managing memory manually. This makes it easier to write Python code without the risk of memory leaks or other memory-related bugs. However, it's still important to be aware of memory usage in large or memory-intensive applications, and to use techniques such as profiling and optimization to ensure that code is running efficiently.

Run the installer and follow the prompts to install Python. Make sure to select the option to add Python to your system PATH during the installation process.

Once the installation is complete, open the Command Prompt by pressing the Windows key + R, typing "cmd", and pressing Enter.

Type "python" in the Command Prompt and press Enter. This will launch the Python interpreter.

To confirm that Python is installed and the PATH variables are set up correctly, type "python --version" in the Command Prompt and press Enter. This should display the version of Python you just installed.

If you encounter any issues, you may need to manually add Python to your PATH variables. To do this, go to Control Panel > System > Advanced system settings > Environment Variables.

Under "System variables", scroll down and find the "Path" variable. Click "Edit" and add the path to the directory where Python is installed (e.g., "C:\Python39"). Make sure to separate the path from any existing entries with a semicolon (;).

Click "OK" to save the changes and close all windows.

Open a new Command Prompt window and type "python" to launch the Python interpreter.
