Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df=pd.read_excel("flight_price.xlsx")

# Extract the flight prices from the dataset
flight_prices = df["Price"]

# Create a histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(flight_prices, bins=20, color='blue', alpha=0.7)
plt.title("Flight Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
# Extract the flight prices from the dataset
flight_prices = df["Price"]

# Calculate the minimum and maximum prices
min_price = flight_prices.min()
max_price = flight_prices.max()

# Calculate the range
price_range = max_price - min_price

print(f"Minimum Price: ${min_price}")
print(f"Maximum Price: ${max_price}")
print(f"Price Range: ${price_range}")
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your app dataset into a DataFrame (replace "your_dataset.csv" with the actual file path)
df = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv")

# Create a boxplot to compare ratings by category
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)  # Rotate category labels for better readability
plt.title("App Ratings by Category")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Use a boxplot to compare ratings by category
plt.boxplot([df[df["Category"] == category]["Rating"] for category in df["Category"].unique()], labels=df["Category"].unique())

plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt



# Convert the "Type" column to strings
df["Type"] = df["Type"].astype(str)

# Filter out rows with specific non-numeric values in the "Type" column
df = df[~df["Type"].str.contains("Everyone")]

# Convert the "Price" column to numeric (if it's not already)
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Group the data by app type and calculate the average price
average_prices_by_type = df.groupby("Type")["Price"].mean()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(8, 6))
average_prices_by_type.plot(kind="bar", color="skyblue")
plt.title("Average Prices by App Type")
plt.xlabel("App Type")
plt.ylabel("Average Price")
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
