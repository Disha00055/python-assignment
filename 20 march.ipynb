```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# Assuming the dataset is in a CSV file named 'customer_data.csv'
df = pd.read_csv('taxonomy.csv')
```

```python
categorical_features = ['name', 'parent_name']
numerical_features = ['taxonomy_id', 'parent_id']
```

```python
# Create an instance of the OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the categorical features
encoded_categorical = encoder.fit_transform(df[categorical_features])

# Create a DataFrame with the encoded categorical features
df_encoded_categorical = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(categorical_features))

# Concatenate the encoded categorical features with the original DataFrame
encoded_df = pd.concat([df_encoded_categorical, df[numerical_features]], axis=1)
```

```python
pd.concat([df,encoded_df],axis=1)
```
